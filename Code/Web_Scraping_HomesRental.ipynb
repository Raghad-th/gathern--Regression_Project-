## Web Scraping using Selenium and BeautifulSoup

```python
main = "https://gathern.co"
chromedriver = "/Users/raghadalthunayan/Desktop/T5_Bootcamp/chromedriver"
os.environ["webdriver.chrome.driver"] = chromedriver

driver = webdriver.Chrome(chromedriver)
driver.get(main)

time.sleep(5)

city = driver.find_element_by_xpath('/html/body/div[1]/main/div[1]/div[1]/div/div/div[1]/div/label/div/div')
city.click()

time.sleep(5)

all_city = driver.find_element_by_xpath('//*[@id="small_selector"]/div[2]/h3')
all_city.click()

time.sleep(5)

type_ = driver.find_element_by_xpath('/html/body/div[1]/main/div[1]/div[1]/div/div/div[3]/div[1]/div/div/div[1]')
type_.click()

time.sleep(5)

all_type = driver.find_element_by_xpath('//*[@id="small_selector"]/div[3]/h3[1]')
all_type.click()

time.sleep(5)

proceed = driver.find_element_by_xpath('/html/body/div[1]/main/div[1]/div[1]/div/div/div[3]/div[3]/div/div/div[1]')
proceed.click()

time.sleep(5)

search = driver.find_element_by_xpath('//*[@id="model_btn2_continue"]')
search.click()

time.sleep(5)


test = False
if test:
    last_page = driver.find_element_by_xpath('//*[@id="mainresultscontainer"]/div/div[2]/ul/li[8]')
    last_page.click()
    
    time.sleep(10)

link_list = []

while True:
    time.sleep(5)
    
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    info_list = soup.find('div', class_="items-lists").find_all('div', class_='ChaletItemBox__content')
    
    
    for i in range(len(info_list)):
        link = 'https://gathern.co'+info_list[i].h3.a['href']
        link_list.append(link)
        
    print('total number of links {}'.format(len(link_list)))
    
    try:
        button = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="mainresultscontainer"]/div/div[2]/button[2]')))
        #button = driver.find_element_by_xpath('//*[@id="mainresultscontainer"]/div/div[2]/button[2]')
        button.click()
        time.sleep(5)
    except:
        print('last page reached')
        break
```

```python
with open('links.txt', 'w') as file:
    
    for i in range(len(link_list)):
        file.write(link_list[i]+'\n')
```
