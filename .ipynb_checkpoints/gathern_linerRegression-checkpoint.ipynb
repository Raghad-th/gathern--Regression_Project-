```python
#read file links 
links_list=[]

with open('links.txt') as file:
    for l in file:
        links_list.append(l.strip())
        
          
links_list
```

```python

Data= pd.DataFrame(data=None ,columns= ['property','city','area','category','rating', 'price'])

for i in range(len(links_list)):
    
    response = requests.get(links_list[i])
    soup = BeautifulSoup(response.text, 'html5lib')

    try:
        Property= soup.find('h1',class_='headTitle').text #get element
    except:
        Property = None 
    
    p_list= soup.find_all('p')
    
    city=p_list[1].text
    
    area=p_list[2].text
    
    category=p_list[3].text
    
    rating=p_list[0].text.strip().split()[0]
    
    price= soup.find_all('b')[-1].text
    
    Data.loc[i] = [Property, city, area, category, rating, price]
    
```

```python
""""
list_p=[]
list_c=[]
list_a=[]
list_ca=[]
list_r=[]
list_pr=[]

for i in range(len(links_list)):
    response = requests.get(links_list[i])
    
    soup = BeautifulSoup(response.text, 'html5lib')

    Property= soup.find('h1',class_='headTitle').text #get element 
    list_p.append(Property)
    
    p_list= soup.find_all('p')
    
    city=p_list[1].text
    list_c.append(city)
    
    area=p_list[2].text
    list_a.append(area)
    
    category=p_list[3].text
    list_ca.append(category)
    
    rating=p_list[0].text.strip().split()[0]
    list_r.append(rating)
    
    price= soup.find_all('b')[-1].text  
    list_pr.append(price)
    
    
new_dict={'property':list_p ,'city':list_c , 'area': list_a ,'category': list_ca , 'rating':list_r ,'price':list_pr}

new_dict=df = pd.DataFrame ({'property':list_p ,'city':list_c , 'area': list_a ,'category': list_ca , 'rating':list_r ,'price':list_pr})
new_dict
""""
```
