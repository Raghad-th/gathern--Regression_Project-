```python
#read file links 
links_list=[]

with open('links.txt') as file:
    for l in file:
        links_list.append(l.strip())
        
          
links_list
```

```python
Data= pd.DataFrame(data=None ,columns= ['property','city','area','category','rating', 'price'])

for i in range(len(links_list)):
    
    response = requests.get(links_list[i])
    soup = BeautifulSoup(response.text, 'html5lib')

    try:
        Property= soup.find('h1',class_='headTitle').text #get element
    except:
        Property = None 
    
    p_list= soup.find_all('p')
    
    city=p_list[1].text
    
    area=p_list[2].text
    
    category=p_list[3].text
    
    rating=p_list[0].text.strip().split()[0]
    
    price= soup.find_all('b')[-1].text
    
    Data.loc[i] = [Property, city, area, category, rating, price]
    
```
